```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv("playground-data.csv")

playground_in_neighborhood = dict()
neighborhood_playground_name = dict()

for row in range(data.shape[0]):
    neighborhood = data.iloc[row,7]
    playground_name = data.iloc[row,1]
    if neighborhood not in playground_in_neighborhood:
        playground_in_neighborhood[neighborhood] = 1                    #Keep count of the playgrounds in each neighborhood
        neighborhood_playground_name[neighborhood] = [playground_name]  #Map each neighborhood to a list of their playgrounds
    else:
        playground_in_neighborhood[neighborhood] += 1
        neighborhood_playground_name[neighborhood].append( playground_name )


print(neighborhood_playground_name)
print('\n')
print(playground_in_neighborhood)    
```

```python
plt.figure(figsize=(5,20))          #Set the size/width of the figure wider         
plt.style.use('classic')            #Set the style of the bar chart to classic
plt.barh(list(playground_in_neighborhood.keys()), list(playground_in_neighborhood.values())) #Create a horizontal bar chart of the neighborhoods and the number of playgrounds
plt.title("Playgrounds in Pittsburgh Neighborhood")
plt.ylabel("Neighborhood")
plt.xlabel("Number of Playgrounds")
```

```python
#Create a data frame of Pittsburgh neighborhoods and the number of playgrounds in each neighborhood
data_subset = pd.DataFrame(list(playground_in_neighborhood.values()),index=list(playground_in_neighborhood.keys()),columns=['Count'])
pd.set_option('display.max_colwidth', 250)                      #Set maximum column width to fit all data
data_subset.insert(1, 'Playgrounds', neighborhood_playground_name.values()) #Insert a column for the names of the playgrounds
data_subset.head(20)
```
