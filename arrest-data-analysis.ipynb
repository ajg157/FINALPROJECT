```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv("arrest-data.csv")
data.dropna(inplace=True)

arrest_neighborhood = dict()
offense = dict()

for row in range(data.shape[0]):
    INCIDENTNEIGHBORHOOD = data.iloc[row,9]
    crime = data.iloc[row,7]
    if INCIDENTNEIGHBORHOOD not in arrest_neighborhood:
        arrest_neighborhood[INCIDENTNEIGHBORHOOD] = 1                    #Keep count of the playgrounds in each neighborhood
        offense[INCIDENTNEIGHBORHOOD] = [crime]  #Map each neighborhood to a list of their playgrounds
    else:
        arrest_neighborhood[INCIDENTNEIGHBORHOOD] += 1
        offense[INCIDENTNEIGHBORHOOD].append(crime)

arrest_neighborhood  
```

```python
plt.figure(figsize=(20,40))          #Set the size/width of the figure wider         
plt.style.use('classic')            #Set the style of the bar chart to classic
plt.barh(list(arrest_neighborhood.keys()), list(arrest_neighborhood.values())) #Create a horizontal bar chart of the neighborhoods and the number of playgrounds
plt.title("Arrests in Pittsburgh Neighborhood")
plt.ylabel("Neighborhood")
plt.xlabel("Number of Arrests")
```
