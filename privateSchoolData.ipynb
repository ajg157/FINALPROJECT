```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("Allegheny_County_Private_Schools_Locations.csv")
data.dropna(inplace=True)

fdddd = (data.iloc[:,9]);
counts = (fdddd.value_counts())
plt.title("Private schools in Pittsburgh")
plt.ylabel("Numbers of schools")
plt.xlabel("Neighborhoods")
plt.plot(counts)
```
