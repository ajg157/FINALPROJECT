```python
import pandas as pd
import matplotlib.pyplot as plt
import operator
import collections

%matplotlib inline
```

Serious Crimes and Their Score Weights

- murder: 5  
- voluntary manslaughter: 5
- kidnapping: 4  
- aggravated assault: 3 
- recklessly endangering another person: 3
- endangering welfare of children: 2
- terroristic threats: 2
- others: 1

```python
data = pd.read_csv("arrest-data.csv")
data.dropna(inplace=True)

arrest_neighborhood = dict()
offense = dict()
offense_types = []

for row in range(data.shape[0]):
    INCIDENTNEIGHBORHOOD = data.iloc[row,9]
    crime = data.iloc[row,7]
    crime = crime.lower()
    if crime not in offense_types:
        offense_types.append(crime)
    if INCIDENTNEIGHBORHOOD not in arrest_neighborhood:
        arrest_neighborhood[INCIDENTNEIGHBORHOOD] = 1                    #Keep count of the arrests in each neighborhood
        offense[INCIDENTNEIGHBORHOOD] = [crime]  #Map each neighborhood to a list of their arrests
    else:
        arrest_neighborhood[INCIDENTNEIGHBORHOOD] += 1
        offense[INCIDENTNEIGHBORHOOD].append(crime)

arrest_neighborhood = sorted(arrest_neighborhood.items(), key=operator.itemgetter(1)) #sort by the number of playgrounds from lowest to highest

arrest_neighborhood = collections.OrderedDict(arrest_neighborhood)  #ouput as a dictionary
```

```python
plt.figure(figsize=(20,40))          #Set the size/width of the figure wider         
plt.style.use('classic')            #Set the style of the bar chart to classic
plt.barh(list(arrest_neighborhood.keys()), list(arrest_neighborhood.values())) #Create a horizontal bar chart of the neighborhoods and the number of playgrounds
plt.title("Arrests in Pittsburgh Neighborhood")
plt.ylabel("Neighborhood")
plt.xlabel("Number of Arrests")
```

```python
data = pd.read_csv("CityOfPittsburgh-PlayGround.csv")

playground_in_neighborhood = dict()
neighborhood_playground_name = dict()

for row in range(data.shape[0]):
    neighborhood = data.iloc[row,7]
    playground_name = data.iloc[row,1]
    if neighborhood not in playground_in_neighborhood:
        playground_in_neighborhood[neighborhood] = 1                    #Keep count of the playgrounds in each neighborhood
        neighborhood_playground_name[neighborhood] = [playground_name]  #Map each neighborhood to a list of their playgrounds
    else:
        playground_in_neighborhood[neighborhood] += 1
        neighborhood_playground_name[neighborhood].append( playground_name )

playground_in_neighborhood = sorted(playground_in_neighborhood.items(), key=operator.itemgetter(1)) #sort by the number of playgrounds from lowest to highest

playground_in_neighborhood = collections.OrderedDict(playground_in_neighborhood)  #ouput as a dictionary
```

```python
plt.figure(figsize=(5,20))          #Set the size/width of the figure wider         
plt.style.use('classic')            #Set the style of the bar chart to classic
plt.barh(list(playground_in_neighborhood.keys()), list(playground_in_neighborhood.values())) #Create a horizontal bar chart of the neighborhoods and the number of playgrounds
plt.title("Playgrounds in Pittsburgh Neighborhood")
plt.ylabel("Neighborhood")
plt.xlabel("Number of Playgrounds")
```

```python
data = pd.read_csv("Allegheny_County_Private_Schools_Locations.csv")

private_school_neighborhood = dict()

for row in range(data.shape[0]):
    neighborhood = data.iloc[row,9]
    if neighborhood not in private_school_neighborhood:
        private_school_neighborhood[neighborhood] = 1                    #Keep count of the playgrounds in each neighborhood
    else:
        private_school_neighborhood[neighborhood] += 1

private_school_neighborhood = sorted(private_school_neighborhood.items(), key=operator.itemgetter(1)) #sort by the number of playgrounds from lowest to highest

private_school_neighborhood = collections.OrderedDict(private_school_neighborhood)  #ouput as a dictionary

private_school_neighborhood
```

```python
data = pd.read_csv("Allegheny_County_Public_Schools__Local_Education_Agency_LEAs_Locations.csv")

public_school_neighborhood = dict()

for row in range(data.shape[0]):
    neighborhood = data.iloc[row,14]
    if neighborhood not in public_school_neighborhood:
        public_school_neighborhood[neighborhood] = 1                    #Keep count of the playgrounds in each neighborhood
    else:
        public_school_neighborhood[neighborhood] += 1

public_school_neighborhood = sorted(public_school_neighborhood.items(), key=operator.itemgetter(1)) #sort by the number of playgrounds from lowest to highest

public_school_neighborhood = collections.OrderedDict(public_school_neighborhood)  #ouput as a dictionary

public_school_neighborhood
```

```python
# Score each neighborhood

neighborhood_scores = dict()


# Score based on playgrounds
for neighborhood, count in playground_in_neighborhood.items():
    neighborhood_scores[neighborhood] = count + 2.0
    
# Score based on number of private schools
for neighborhood, count in private_school_neighborhood.items():
    if neighborhood not in neighborhood_scores:
        neighborhood_scores[neighborhood] = count * 3.0
    else:        
        neighborhood_scores[neighborhood] += count * 3.0

# Score based on number of public schools
for neighborhood, count in public_school_neighborhood.items():
    if neighborhood not in neighborhood_scores:
        neighborhood_scores[neighborhood] = count * 1.0
    else:
        neighborhood_scores[neighborhood] += count * 1.0 
    
# Score based on arrests
for neighborhood, crime_list in offense.items():
    arrest_score = 0
    for crime in crime_list:    
        if 'murder' in crime:
            arrest_score += 5.0
        elif 'voluntary manslaughter' in crime:
            arrest_score += 5.0
        elif 'kidnapping' in crime:
            arrest_score += 4.0
        elif 'aggravated assault' in crime:
            arrest_score += 3.0
        elif 'recklessy endangering another person' in crime:
            arrest_score += 3.0
        elif 'endangering welfare of children' in crime:
            arrest_score += 2.0
        elif 'terroristic threat' in crime:
            arrest_score += 2.0
        else:
            arrest_score += 1.0
    
    if neighborhood in neighborhood_scores:
        neighborhood_scores[neighborhood] /= arrest_score * 1000.0
    else:
        neighborhood_scores[neighborhood] = 1.0 / arrest_score * 1000.0

neighborhood_scores = sorted(neighborhood_scores.items(), key=operator.itemgetter(1))
neighborhood_scores = collections.OrderedDict(neighborhood_scores)

# Remove unrelevant cities
del neighborhood_scores['Outside County']
del neighborhood_scores['Outside State']
del neighborhood_scores['Outside City']
del neighborhood_scores['Pittsburgh']

neighborhood_scores  
```

```python
plt.figure(figsize=(20,40))          #Set the size/width of the figure wider         
plt.style.use('classic')            #Set the style of the bar chart to classic
plt.barh(list(neighborhood_scores.keys()), list(neighborhood_scores.values())) #Create a horizontal bar chart of the neighborhoods and the number of playgrounds
plt.xscale('log')   #display the graph in logarithmic scale
plt.title("Best Neighborhood in Pittsburgh")
plt.ylabel("Neighborhood")
plt.xlabel("Score")
```
